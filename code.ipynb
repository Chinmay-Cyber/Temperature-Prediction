```python
import pandas as pd

data = pd.read_csv("data.csv", encoding="latin-1")

print(data.head())
```

```python
data=data.drop(columns=["Area Code","Months Code","Element Code","Unit"])
print(data.head())
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum
```

```python
data_long = data.melt(
    id_vars=["Area","Months","Element"],
    var_name="Year",
    value_name="Temperature"
)
# melt reshapes the dataset we made a new dataset here named data_long
```

```python
print(data_long.head())
```

```python
data_long['Year']=data_long['Year'].str.replace("Y","").astype(int)
```

```python
print(data_long.head())
```

```python
data_long.isnull().sum
data_long=data_long.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
le_area=LabelEncoder()
le_month=LabelEncoder()

data_long["Area"]=le_area.fit_transform(data_long["Area"])
data_long["Months"]=le_month.fit_transform(data_long["Months"])
```

```python
X=data_long[['Area','Months','Year']]
Y=data_long['Temperature']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
# train the  model 
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100)
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
prediction=model.predict(X_test)
mae=mean_absolute_error(y_test,prediction)
print(mae)
```
